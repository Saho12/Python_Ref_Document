```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
%matplotlib qt
%matplotlib inline
```

```python
t= np.arange(0,2*3.14,0.1)
s1 = np.sin(t)
s2 = np.cos(t) 
s3 = np.sin(t) +np.cos(t)
```

```python
fig1 = plt.figure(figsize=(8,5)) 
fig1.suptitle('Sharing x per column, y per row')
ax1 = plt.subplot(3,2,1)
plot_1 = ax1.plot(t, s1, 'r*--', label= 'Sinusoid')
ax1.legend()
ax1.set_xlabel('Time')
ax1.set_ylabel('Sinus')
ax1.set_title('Sinus')

ax2= plt.subplot(3,2,2)
plot_2 = ax2.plot(t, s2, 'b*--', label= 'CoSinusoid')
ax2.legend()
ax2.set_xlabel('Time')
ax2.set_ylabel('CoSinus')
ax2.grid()
ax2.set_title('Cosinus')


ax3= plt.subplot(3,1,2)
plot_2 = ax3.plot(t, s3, 'b*-', label= 'CoSin')
ax3.legend()
ax3.set_xlabel('Time')
ax3.set_ylabel('CoSin')
ax3.set_title('Sumation')
ax3.set_title('Both')
#plt.show()
plt.savefig('Demo.pdf', format= 'pdf')
```

### BarPlot

```python
data=pd.read_excel('summary.xlsx', skiprows=3)
data.set_index('Mean_Abs_diff_Current', inplace=True)
```

```python
# Vertical Bar Plot with value at Top of bar

xx=data.index.values[2:7]
x_index= np.arange(len(xx))
y_index= np.arange(0,4,0.5)
width=0.25
ii=0 

def inser_data_labels(bars):
    for bar in bars:
        bar_height= bar.get_height()
        ax.annotate('{0:.1f}'.format (bar.get_height()),
                    xy= (bar.get_x()+bar.get_width()/2, bar_height),
                    xytext=(0,3),
                    textcoords= 'offset points',
                    ha='center',
                    va= 'bottom'                    
                   )
        
for i in range(len (data.columns)):
    
    yy=data.iloc[:,i][2:7]
    if(i%3==0):
        fig,ax = plt.subplots()
    curr_bar= ax.bar(x_index+(i%3)*width,yy,label= data.columns[i], width=width)
    inser_data_labels(curr_bar)
    if (ii>=2):
        ax.set_ylabel('Mean_Dev[Ampere]',fontsize='10')
        ax.set_xlabel('Current_Range[Ampere]',fontsize='10')
        ax.set_xticks(ticks= x_index)
        ax.set_xticklabels(labels=xx)
        ax.set_yticks(ticks= y_index)
        
        plt.legend()
        plt.savefig('Mean_AbsDev_Plot'+data.columns[i].split('_')[2],format='pdf',dpi=900)
        plt.close()
        ii=0
        plt.show()
    else:
        ii+=1
```

### Axis Font size change

```python
import matplotlib.pyplot as plt

# type-1

SMALL_SIZE = 8
MEDIUM_SIZE = 10
BIGGER_SIZE = 12

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=SMALL_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
 
# type-2  calling the rc method on matplotlib:

import matplotlib

SMALL_SIZE = 8
matplotlib.rc('font', size=SMALL_SIZE)
matplotlib.rc('axes', titlesize=SMALL_SIZE)
```
